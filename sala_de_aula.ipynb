```python
Questão em Go para determinar o tipo de triângulo
package main

import "fmt"

func main() {

	var a, b, c float64

	fmt.Println("Escreva os valores númericos dos três lados do triângulo:")
	fmt.Scan(&a, &b, &c)

	if a < b+c && b < a+c && c < a+b {
		if a == b && a == c && b == c {
			fmt.Println("O triângulo é equilátero")
		} else if a == b && a != c || a == c && a != b || c == b && c != a {
			fmt.Println("O triângulo é isósceles")
		} else {
			fmt.Println("O triângulo é escaleno")
		}
	} else {
		fmt.Println("Os valores inseridos não formam um triângulo")
	}
}
```

```python
Questão em GO sobre determinar a média das notas ponderadas, e falar se o aluno esta aprovado ou reprovado
package main

import "fmt"

func main() {

	var nota1, nota2, peso1, peso2, media float64
	fmt.Print("Escreva a nota 1 e seu peso, respectivamente:")
	fmt.Scan(&nota1, &peso1)

	fmt.Print("Escreva a nota 2 e seu peso, respectivamente:")
	fmt.Scan(&nota2, &peso2)

	media = (nota1*peso1 + nota2*peso2) / (peso1 + peso2)

	fmt.Printf("A média ponderada é: %.2f\n", media)

	if media >= 7 {
		fmt.Println("Como a média é maior ou igual a 7, o aluno esta aprovado.")
	} else {
		fmt.Println("Como a média é menor que 7, o aluno esta reprovado.")
	}
}
```
