```python
algoritmo "composicao"
var
   n1, n2, n3, numero: inteiro
inicio
   escreva("Escreva 3 números:")
   leia(n1, n2, n3)

   se n1 > 9 ou n2 > 9 ou n3 > 9 entao
      escreva("DIGITO INVALIDO\n")
   senao
      numero <- n1*100 + n2*10 + n3
      escreva(numero, ", ", numero*numero,)
   fimse
fimalgoritmo
```

```python
algoritmo "energia"
var
   salario, kw, valor_kw, total, desconto: real
inicio
   escreva("Digite o valor do salário e a quantidade de kw gasto na residência:")
   leia(salario, kw)

   valor_kw <- (salario * 0.7) / 100
   total <- valor_kw * kw
   desconto <- total * 0.9

   escreval("Custo por kW: R$ ", valor_kw)
   escreval("Custo do consumo: R$ ", total:)
   escreva("Custo com desconto: R$ ", desconto)
fimalgoritmo
```

```python
algoritmo "agua"
var
   conta: inteiro
   consumo: real
   tipo: caractere
   valor: real
inicio
   escreva("Digite o identificador da conta(C-consumidor, I-industrial e R-residencial), escreva o consumo e a categoria de consumidor, respectivamente:")
   leia(conta, consumo, tipo)

   se tipo = "R" entao
      valor <- 5 + consumo * 0.05
   senao
      se tipo = "C" entao
         valor <- 500 + consumo * 0.25
      senao
         valor <- 800 + consumo * 0.04
      fimse
   fimse

   escreva("CONTA = ", conta)
   escreva("VALOR DA CONTA = ", valor:0:2)
fimalgoritmo
```

```python
algoritmo "temp"
var
   n, i: inteiro
   f, c: real
inicio
   leia(n)

   para i de 1 ate n faca
      leia(f)
      c <- (5*(f-32))/9
      escreva(f:0:2, " FAHRENHEIT EQUIVALE A ", c:0:2, " CELSIUS\n")
   fimpara
fimalgoritmo
```

```python
algoritmo "metrico"
var
   f, c, polegadas, mm: real
inicio
   leia(f)
   leia(polegadas)

   c <- (5*f - 160)/9
   mm <- polegadas * 25.4

   escreva("O VALOR EM CELSIUS = ", c:0:2, "\n")
   escreva("A QUANTIDADE DE CHUVA E = ", mm:0:2, "\n")
fimalgoritmo
```

```python
algoritmo "lata"
var
   r, h, area, custo: real
inicio
   leia(r)
   leia(h)

   area <- 2*3.14159*r*r + 2*3.14159*r*h
   custo <- area * 100

   escreva("O VALOR DO CUSTO E = ", custo:0:2, "\n")
fimalgoritmo
```

```python
algoritmo "delta"
var
   a, b, c, d: real
inicio
   leia(a)
   leia(b)
   leia(c)

   d <- b*b - 4*a*c

   escreva("O VALOR DE DELTA E = ", d:0:2, "\n")
fimalgoritmo
```

```python
algoritmo "det"
var
   a, b, c, d, det: real
inicio
   leia(a)
   leia(b)
   leia(c)
   leia(d)

   det <- a*d - b*c

   escreva("O VALOR DO DETERMINANTE E = ", det:0:2, "\n")
fimalgoritmo
```

```python
algoritmo "divisivel"
var
   n: inteiro
inicio
   leia(n)

   se n mod 3 = 0 e n mod 5 = 0 entao
      escreva("O NUMERO E DIVISIVEL\n")
   senao
      escreva("O NUMERO NAO E DIVISIVEL\n")
   fimse
fimalgoritmo
```

```python
algoritmo "charrete"
var
   h: inteiro
   valor: real
inicio
   leia(h)

   valor <- (h div 3) * 10 + (h mod 3) * 5

   escreva("O VALOR A PAGAR E = ", valor:0:2, "\n")
fimalgoritmo
```

```python
algoritmo "conceito"
var
   nota: real
inicio
   leia(nota)

   se nota >= 9 entao
      escreva("NOTA = ", nota:0:1, " CONCEITO = A\n")
   senao se nota >= 7.5 entao
      escreva("NOTA = ", nota:0:1, " CONCEITO = B\n")
   senao se nota >= 6 entao
      escreva("NOTA = ", nota:0:1, " CONCEITO = C\n")
   senao
      escreva("NOTA = ", nota:0:1, " CONCEITO = D\n")
   fimse
fimalgoritmo
```

```python
algoritmo "piramide"
var
   h, a, area, volume: real
inicio
   leia(h)
   leia(a)

   area <- (3*a*a*raizq(3))/2
   volume <- (area*h)/3

   escreva("O VOLUME DA PIRAMIDE E = ", volume:0:2, " METROS CUBICOS\n")
fimalgoritmo
```

```python
algoritmo "pares"
var
   n, i: inteiro
inicio
   leia(n)

   para i de 2 ate n passo 2 faca
      escreva(i, "^2 = ", i*i, "\n")
   fimpara
fimalgoritmo
```

```python
algoritmo "reajuste"
var
   s: real
inicio
   leia(s)

   se s <= 300 entao
      s <- s * 1.5
   senao
      s <- s * 1.3
   fimse

   escreva("SALARIO COM REAJUSTE = ", s:0:2, "\n")
fimalgoritmo
```

```python
algoritmo "serie"
var
   x, y, i: inteiro
inicio
   leia(x, y)

   se x mod 2 <> 0 entao
      escreva("O PRIMEIRO NUMERO NAO E PAR\n")
   senao
      para i de 0 ate y-1 faca
         escreva(x + i*2, " ")
      fimpara
      escreva("\n")
   fimse
fimalgoritmo
```

```python
algoritmo "pa"
var
   a1, r, n, i, soma, termo: inteiro
inicio
   leia(a1, r, n)

   soma <- 0
   termo <- a1

   para i de 1 ate n faca
      soma <- soma + termo
      termo <- termo + r
   fimpara

   escreva(soma, "\n")
fimalgoritmo
```

```python
algoritmo "somatorio"
var
   n, i: inteiro
   s: real
inicio
   leia(n)

   se n <= 1 entao
      escreva("Numero invalido!\n")
   senao
      s <- 0
      para i de 1 ate n faca
         s <- s + 1/i
      fimpara
      escreva(s:0:6, "\n")
   fimse
fimalgoritmo
```
