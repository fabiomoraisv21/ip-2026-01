```python
Algoritmo "Arrecadação_de_Jogos"

Var

    total: inteiro
    pop, ger, arq, cad, arrecadacao: real

Inicio

    escreva("Qual o número total de pessoas que iram assistir ao jogo?")
    leia(total)

    escreval("Qual a porcetagem das categorias? Escreva, respectivamente, popular, geral, arquibancada e cadeiras")
    leia(pop, ger, arq, cad)

    arrecadacao <- total / 100 * (pop + ger * 5 + arq * 10 + cad * 20)

    Escreva("A arrecadação total do jogo será de R$", arrecadacao)

Fimalgoritmo
```

```python
algoritmo "composicao"
var
   n1, n2, n3, numero: inteiro
inicio
   escreva("Escreva 3 números:")
   leia(n1, n2, n3)

   se n1 > 9 ou n2 > 9 ou n3 > 9 entao
      escreva("DIGITO INVALIDO\n")
   senao
      numero <- n1*100 + n2*10 + n3
      escreva(numero, ", ", numero*numero,)
   fimse
fimalgoritmo
```

```python
algoritmo "energia"
var
   salario, kw, valor_kw, total, desconto: real
inicio
   escreva("Digite o valor do salário e a quantidade de kw gasto na residência:")
   leia(salario, kw)

   valor_kw <- (salario * 0.7) / 100
   total <- valor_kw * kw
   desconto <- total * 0.9

   escreval("Custo por kW: R$ ", valor_kw)
   escreval("Custo do consumo: R$ ", total:)
   escreva("Custo com desconto: R$ ", desconto)
fimalgoritmo
```

```python
algoritmo "agua"
var
   conta: inteiro
   consumo: real
   tipo: caractere
   valor: real
inicio
   escreva("Digite o identificador da conta, escreva o consumo e a categoria de consumidor (C-consumidor, I-industrial e R-residencial), respectivamente:")
   leia(conta, consumo, tipo)

   se tipo = "R" entao
      valor <- 5 + consumo * 0.05
   senao
      se tipo = "C" entao
         valor <- 500 + consumo * 0.25
      senao
         valor <- 800 + consumo * 0.04
      fimse
   fimse

   escreva("CONTA = ", conta)
   escreva("VALOR DA CONTA = ", valor)
fimalgoritmo
```

```python
algoritmo "ConversaoTemperatura"

var
   n, i: inteiro
   F, C: real

inicio
   escreva("Digite a quantidade de temperaturas: ")
   leia(n)

   para i de 1 ate n faca
      escreva("Digite a ", i, "ª temperatura em Fahrenheit: ")
      leia(F)

      C <- (5 * (F - 32)) / 9

      escreval(F:0:2, " FAHRENHEIT EQUIVALE A ", C, " CELSIUS")
   fimpara

fimalgoritmo
```

```python
algoritmo "Metrico"

var
   f, polegada, c, mm: real

inicio
   escreva("escreva a temperatura em Fahrenheit:")
   leia(f)
   escreva("Digite a quantidade de chuva em polegadas")
   leia(polegada)

   c <- (5 * (f - 32))/9
   mm <- polegada * 25.4

   escreval("O VALOR EM CELSIUS = ", c)
   escreval("A QUANTIDADE DE CHUVA E = ", mm)

fimalgoritmo
```

```python
lgoritmo "Lata"

var
   r, h, area, custo: real
   pi: real

inicio

   pi <- 3.14159

   escreva("Digite o raio da lata (em metros): ")
   leia(r)

   escreva("Digite a altura da lata (em metros): ")
   leia(h)

   area <- 2*pi*r*r + 2*pi*r*h
   custo <- area * 100

   escreval("O VALOR DO CUSTO E = ", custo)
```

```python
algoritmo "Delta"

var
   a, b, c, delta: real

inicio

   escreva("Digite o valor de A: ")
   leia(a)

   escreva("Digite o valor de B: ")
   leia(b)

   escreva("Digite o valor de C: ")
   leia(c)

   delta <- b*b - 4*a*c

   escreval("O VALOR DE DELTA E = ", delta)

fimalgoritmo
```

```python
algoritmo "Determinante"

var
   a, b, c, d, det: real

inicio

   escreva("Digite o valor de a: ")
   leia(a)

   escreva("Digite o valor de b: ")
   leia(b)

   escreva("Digite o valor de c: ")
   leia(c)

   escreva("Digite o valor de d: ")
   leia(d)

   det <- a*d - b*c

   escreval("O VALOR DO DETERMINANTE E = ", det)

fimalgoritmo
```

```python
algoritmo "Divisivel"

var
   n: inteiro

inicio

   escreva("Digite um numero inteiro: ")
   leia(n)

   se n % 3 = 0 e n % 5 = 0 entao
      escreval("Resultado: O NUMERO E DIVISIVEL POR 3 E 5")
   senao
      escreval("Resultado: O NUMERO NAO E DIVISIVEL POR 3 E 5")
   fimse

fimalgoritmo
```

```python
algoritmo "Charrete"

var
   h: inteiro
   valor: real

inicio

   escreva("Digite a quantidade de horas de uso: ")
   leia(h)

   valor <- (h / 3)*10 + (h % 3)*5

   escreval("O VALOR A PAGAR E = R$ ", valor)

fimalgoritmo
```

```python
algoritmo "Conceito"

var
   nota: real
   conceito: caractere

inicio

   escreva("Digite a nota (0 a 10): ")
   leia(nota)

   se nota >= 9 entao
      conceito <- "A"
   senao se nota >= 7.5 entao
      conceito <- "B"
   senao se nota >= 6 entao
      conceito <- "C"
   senao
      conceito <- "D"
   fimse

   escreval("NOTA = ", nota, " | CONCEITO = ", conceito)

fimalgoritmo
```

```python
algoritmo "Piramide"

var
   h, a, ab, volume: real

inicio

   escreva("Digite a altura da piramide (m): ")
   leia(h)

   escreva("Digite o lado do hexagono (m): ")
   leia(a)

   ab <- (3*a*a*raiz(3))/2
   volume <- (ab*h)/3

   escreval("VOLUME = ", volume, " METROS CUBICOS")

fimalgoritmo
```

```python
algoritmo "QuadradoPares"

var
   n, i: inteiro

inicio

   escreva("Digite um numero inteiro (5 < N < 2000): ")
   leia(n)

   se (n <= 5) ou (n >= 2000) entao
      escreval("Valor invalido!")
   senao

      escreval("Resultados:")

      para i de 2 ate n passo 2 faca
         escreval(i, "^2 = ", i*i)
      fimpara

   fimse

fimalgoritmo
```

```python
algoritmo "Reajuste"

var
   salario, novo: real

inicio

   escreva("Digite o salario atual: ")
   leia(salario)

   se (salario <= 300) entao
      novo <- salario * 1.5
   senao
      novo <- salario * 1.3
   fimse

   escreval("SALARIO COM REAJUSTE = R$ ", novo)

fimalgoritmo
```

```python
algoritmo "SeriePares"

var
   x, y, i: inteiro

inicio

   escreva("Digite o primeiro numero: ")
   leia(x)

   escreva("Digite a quantidade de termos: ")
   leia(y)

   se x % 2 <> 0 entao
      escreval("O PRIMEIRO NUMERO NAO E PAR")
   senao

      escreval("Serie gerada:")

      para i de 0 ate y-1 faca
         escreva(x + i*2, " ")
      fimpara

   fimse

fimalgoritmo
```

```python
algoritmo "PA"

var
   a1, r, n, i, soma, termo: inteiro

inicio

   escreva("Digite o primeiro termo: ")
   leia(a1)

   escreva("Digite a razao: ")
   leia(r)

   escreva("Digite a quantidade de termos: ")
   leia(n)

   soma <- 0

   para i de 0 ate n-1 faca
      termo <- a1 + i*r
      soma <- soma + termo
   fimpara

   escreval("SOMA = ", soma)

fimalgoritmo
```

```python
algoritmo "Somatorio"

var
   n, i: inteiro
   s: real

inicio

   escreva("Digite um numero inteiro maior que 1: ")
   leia(n)

   se n <= 1 entao
      escreval("Numero invalido!")
   senao

      s <- 0

      para i de 1 ate n faca
         s <- s + 1/i
      fimpara

      escreval("RESULTADO = ", s)

   fimse

fimalgoritmo
```

```python
algoritmo "Tempo"

var
   h, m, s, total: inteiro

inicio

   escreva("Digite as horas: ")
   leia(h)

   escreva("Digite os minutos: ")
   leia(m)

   escreva("Digite os segundos: ")
   leia(s)

   total <- h*3600 + m*60 + s

   escreval("O TEMPO EM SEGUNDOS E = ", total)

fimalgoritmo
```
